```python
import seaborn as sns
import pandas as pd
```

```python
sns.set(style='darkgrid')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='time')
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri)
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri, ci='sd') #confidence interval(ci)
#or
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, ci='sd', hue='event', kind='line')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, ci='sd', hue='region', kind='line', style='event')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', col='time')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', col='size')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, col='event', hue='region')
```

```python
sns.lineplot(x = 'timepoint', y  = 'signal', style = 'event', hue = 'region', data = fmri, markers = True, err_style='bars')
```
